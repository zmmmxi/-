```python
import csv
import py2neo
from py2neo import Graph,Node,Relationship,NodeMatcher,RelationshipMatcher,NodeMatch
```

```python
g=Graph('http://localhost:7474',user='neo4j',password='12345')
```

```python
with open('./药物疗法.csv','r',encoding='utf-8') as f:
    reader=csv.reader(f)
    therapy=Node("疗法",name='药物疗法')
    g.merge(therapy,"疗法","name")
    for item in reader:
        if reader.line_num==1:
            continue
        print("当前行数：",reader.line_num,"当前内容：",item)
        specialty=Node("专科",name=item[0])
        DiseaseName=Node("病症",name=item[1])
        DiseaseName['简介']=item[2]
        classification=Node("病名",name=item[4])
        prescription=Node("处方名",name=item[3])
        prescription['处方']=item[5]
        prescription['用法']=item[6]
        prescription['疗效']=item[7]
        prescription['病例']=item[8]
        prescription['注释']=item[9]
        relation=Relationship(therapy,'包含专科',specialty)
        relation1=Relationship(specialty,'包含病症',DiseaseName)
        relation2=Relationship(DiseaseName,'包含疾病',classification)
        relation3=Relationship(classification,'包含处方',prescription)
#         relation4=Relationship(prescriptionName,'处方内容',prescription)
        g.merge(specialty,"专科","name")
        g.create(relation)
        g.merge(DiseaseName,"病症","name")
        g.create(relation1)
        g.merge(classification,"病名","name")
        g.create(relation2)
        g.merge(prescription,"处方名","name")
        g.create(relation3)
#         g.merge(prescription,"处方","name")
#         g.merge(relation4)
```

```python
with open('./烧刺疗法.csv','r',encoding='utf-8') as f:
    reader=csv.reader(f)
    therapy=Node("疗法",name='烧刺疗法')
    g.merge(therapy,"疗法","name")
    for item in reader:
        if reader.line_num==1:
            continue
        print("当前行数：",reader.line_num,"当前内容：",item)
        Disease=Node("病名",name=item[0])
        Disease['疗法']=item[1]
        Disease['疗效']=item[2]
        Disease['病例']=item[3]
        relation=Relationship(therapy,'包含疾病',Disease)
        g.merge(Disease,"病名","name")
        g.create(relation)
```

```python
select_name = input("请输入查询的病症名：")
match_disease = g.run("match (q:`病症`{name:'"+select_name+"'}) return q.`简介`").data()
print(match_disease)
```

```python
select_name1 = input("请输入查询的疾病名：")
match_prescription = g.run("match (b:`病名`{name:'"+select_name1+"'})-[m:`包含处方`]-(n) return n.name,n.`处方`").data()
print(match_prescription)
```
